```python
def equal(k,l,m):
  count = 0
  lst = [k,l,m]
  highest_num =  max(set(lst), key = lst.count) 
  for i in lst:
    if i == highest_num:
      count += 1
  if count == 1:
    return 0 # becuse we can have 2 equal value not one
  else:
    return count
```

```python
equal(3,4,3)
```

```python
equal(1,1,1)
```

```python
equal(3,4,1)
```

```python
from collections import OrderedDict
def dict_to_list(dict_):
  return  list(OrderedDict(sorted(dict_.items())).items())
dict_to_list({'D':1,"B":2,"C":3})
```

```python
dict_to_list({"likes":2,"dislikes":3,"followers":10})
```

```python
def mapping(lst):
  return {i:i.upper() for i in lst}

mapping(['p','s'])
```

```python
mapping(['a','b','c'])
```

```python
mapping(['a','v','y','z'])
```

```python
def vow_replace(sent, ch):
  return ''.join(char if char.lower() not in 'aeiou' else ch for char in sent )
```

```python
vow_replace('apples and bananas','u')
```

```python
vow_replace('cheese casserole','o')
```

```python
vow_replace('stuffed jalapeno poppers','e')
```

```python
def ascii_capitalize(sent):
  return ''.join(char.upper() if ord(char)%2==0 else char.lower() for char in sent )
```

```python
ascii_capitalize('to be or not to be!')
```

```python
ascii_capitalize('THE LITTLE MERMAID')
```

```python
ascii_capitalize('oh what a beautiful morning')
```
