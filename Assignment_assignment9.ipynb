```python
# Q.1 
# A disarium number is a number in which the sum of the digits to the power of
# their respective position is equal to the number itself 
# return pow((num%10), p) + sumOfDigits(num/10, p-1);

def calculateSum(num, len):
  if num == 0:
    return 0
  else:
    return (num%10)**len + calculateSum(num//10, len-1)


def IsDisarium(num):
  def calculatelength(num):
    length = 0 
    while num > 0:
      length += 1
      num //= 10
    return length
  length = calculatelength(num)
  sum = calculateSum(num , length)
  return sum == num

IsDisarium(175)
```

```python
# question 3 
# In number theory, a happy number is a number which eventually reaches 1 when
# replaced by the sum of the square of each digit. For instance, 13 is a happy number 
# calculatin perfect digital invariant
def pdi(number):
    total = 0
    while number > 0:
        total += pow(number % 10, 2)
        number = number // 10
    return total

def is_happy(number):
    """Determine if the specified number is a happy number."""
    pdi_numbers = []
    while number > 1 and number not in pdi_numbers:
        pdi_numbers.append(number)
        number = pdi(number)
    #print('all perfect digital invariant->',pdi_numbers)
    return number == 1

is_happy(13)
```

```python
# printing all the happy numbers

def PrintHappy():
  for i in range(1,100):
    if is_happy(i):
      print(i, end = ' ')

PrintHappy()
```

```python
#  harshad number (or Niven number) in a given number base is an integer that 
# is divisible by the sum of its digits when written in that base.
```

```python
def digit_sum(number, base = 10):
    total = 0
    while number > 0:
        total += number % base
        number = number // base
    return total
def isHarshad(number):
  sum = digit_sum(number)
  return number%sum == 0

isHarshad(18)
```

```python
# Question . 6
# A pronic number is a number which is the product of two consecutive integers, 
# that is, a number of the form n(n + 1).
def isPronic(num):
  for i in range(num):
    if num == i * (i+1):
      return True
  return False

isPronic(54)
```

```python
def PrintPronic():
  for i in range(1,100):
    if isPronic(i):
      print(i, end = ' ')

PrintPronic()
```
