```python
def unique(lst):
  return min(set(lst), key = lst.count)
```

```python
unique([3,3,3,7,3,3])
```

```python
unique([0,0,0.77,0,0])
```

```python
unique([0,1,1,1,1,1,1,1])
```

```python
import math
class Circle:
  def __init__(self, radius):
    self.radius = radius
    self.PI = math.pi
  def getArea(self):
    return self.PI * (self.radius)**2
  def getPerimeter(self):
    return 2 * self.PI * self.radius

circy = Circle(11)
circy.getArea()
```

```python
circy = Circle(4.44)
circy.getPerimeter()
```

```python
def is_triplet(k,l,m):
  lst = [k,l,m]
  max_ = max(lst)
  min_ =  min(lst)
  lst.remove(max_)
  lst.remove(min_)
  second_min = lst[0]

  return (min_**2) + (second_min**2) == max_ ** 2


```

```python
is_triplet(3,4,5)
```

```python
is_triplet(13,5,12)
```

```python
is_triplet(1,2,3)
```
