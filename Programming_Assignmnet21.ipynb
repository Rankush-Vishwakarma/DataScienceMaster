```python
def next_in_line(lst,num):
  if len(lst) != 0:
    lst.append(num)
    new = lst[1:]
    return new
  else:
    return 'No list has been selected'

next_in_line([5,6,7,8,9],1)
```

```python
next_in_line([],6)
```

```python
def get_budgets(lst):
  sum_ = 0
  for i in lst:
    sum_ += i['budget']
  return sum_

get_budgets([{'name':'john','age':21,'budget':23000},
             {'name':'steve','age':32,'budget':40000},
             {'name':'martin','age':16,'budget':2700}])
```

```python
get_budgets([{'name':'john','age':21,'budget':29000},
             {'name':'steve','age':32,'budget':32000},
             {'name':'martin','age':16,'budget':1600}])
```

```python
def alphabet_soup(str):
  lst = list(str)
  lst.sort()
  return ''.join(lst)

alphabet_soup('hello')
```

```python
alphabet_soup('edabit')
```

```python
alphabet_soup('hacker')
```

```python
def compund_interest(p,t,r,n):
  return round(p * (1 + (r/n))**(n*t),2)
```

```python
compund_interest(100,1,0.05,1)
```

```python
compund_interest(3500,15,0.1,4)
```

```python
def return_only_integer(lst):
  return list(filter(lambda x: type(x) == int, lst))

return_only_integer([9,2,'space','car','lion',16])
```

```python
return_only_integer(['hello',81,'basketball',123,'fox'])
```
