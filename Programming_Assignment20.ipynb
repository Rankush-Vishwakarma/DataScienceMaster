```python
def filter_list(lst):
  return list(filter(lambda x: type(x) == int, lst))

filter_list([1,2,3,'a','b',4])
```

```python
filter_list(['a',0,'Edabit',1729,'python','1729'])
```

```python
def add_indexs(lst):
  for i in range(len(lst)):
    lst[i] += i
  return lst

add_indexs([0,0,0,0,0])
```

```python
add_indexs([1,2,3,4,5])
```

```python
def cone_volume(h,r):
  return round((1/3)*3.14159*(r**2)*h,2)

cone_volume(3,2)
```

```python
cone_volume(15,6)
```

```python
def missing_num(lst):
  # n * (n+1) /2 sum of n natural number 
  sum_of_all_number = 10 *(10+1) // 2
  return sum_of_all_number - sum(lst)

missing_num([1,2,3,4,6,7,8,9,10])
```

```python
missing_num([7,2,3,6,5,9,1,4,8])
```
