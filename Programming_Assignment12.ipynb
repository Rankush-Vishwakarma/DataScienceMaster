```python
# @ . 1
from itertools import chain
def ExtractUnique(dict_):
  return set(chain(*dict_.values()))

ExtractUnique({'i':[1,2,3,4],
               'am':[2,3,4,5,6],
               'rankush':[12,23,34,45]}
              )
```

```python
# Q. 2
def SumOfAllitems(dict_):
  return sum(set(chain(*dict_.values())))

SumOfAllitems({'i':[8],
               'am':[8],
               'rankush':[2,0,0,1]}
              )
```

```python
# Q. 3
def Merge(d1,d2):
  d3 =  {**d1, **d2}
  return d3

print(Merge({'i':8,'am':8,
               'rankush':2001},{'python programmer':[2,0,0,1]}))
```

```python
# Q. 4
def FlatDict(d):
  return dict(zip(d['dob'], d['rankush vishwakarma']))

FlatDict({'dob' : [8, 8, 2001],
             'rankush vishwakarma' : ['tuesday', 'august', 'best year because i have born']})
```

```python
# Q. 5
from collections import OrderedDict
def Insert(OrderDict,valueDict):
  OrderDict.update(valueDict)
  OrderDict.move_to_end(list(valueDict.keys())[0], last = False)
  return OrderDict

Insert(OrderedDict([('rk', 'guys'), ('vishwakarma', 'python programmer')]), {'job':'desperatly'})
```

```python
from collections import OrderedDict 
  
def checkOrder(input, pattern): 
    temp = OrderedDict.fromkeys(input) 
    ptrlen = 0
    for key,value in temp.items(): 
        if (key == pattern[ptrlen]): 
            ptrlen = ptrlen + 1
        if (ptrlen == (len(pattern))): 
            return 'true'
    return 'false'

checkOrder('rankush vishwakarma','an')
```

```python
def Sort(dict_, by='key'):
  if by == 'key':
    dict_ = sorted(dict_.items())
    return dict_ 
  elif  by == 'value':
    return {k: v for k, v in sorted(dict_.items(), key=lambda item: item[1])}
  else:
    raise TypeError(f'Error in by={by}')


Sort({1: 2, 3: 4, 4: 3, 2: 1, 0: 0}, by='value')
```
