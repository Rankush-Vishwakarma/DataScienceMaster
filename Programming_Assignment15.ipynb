```python
def Question1(n):
  def generator(n):
    for i in range(n):
      if i%5 == 0 and i%7 == 0:
        yield i
  que1 =  generator(n)
  print(*que1, sep = ",")
```

```python
Question1(100)
```

```python
def Question2(n):
  def generator(n):
    for i in range(n+1):
      if i%2 == 0:
        yield i
  que2 =  generator(n)
  print(*que2, sep = ",")
```

```python
Question2(10)
```

```python
def Question3(n):
  fib = [0,1]
  [fib.append(fib[i-2]+fib[i-1])for i in range(2,n)]
  return fib
```

```python
Question3(7)
```

```python
class Shape:
  def area(self):
    return self.area
class Square(Shape):
  def __init__(self, length = 0):
    self.length = length 
  def area(self):
    return self.length

s = Square()
s.area()
```
