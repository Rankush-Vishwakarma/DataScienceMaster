in python there are two type of copy exits.

1. Shallow copy (copy.copy())
2. Deep copy (copy.deepcopy())


1. A shallow copy means making a new collection object and then populating it with references to the child objects found in the original.any changes made to a copy of object ** do reflect ** in the original object. In python, this is implemented using “copy()” function. The copying process does not recurse and therefore won’t create copies of the child objects themselves. In case of shallow copy, a reference of object is copied in other object.

2. Deep copy is a process in which the copying process occurs recursively.any changes made to a copy of object ** do not reflect ** in the original object. In python, this is implemented using “deepcopy()” function. It means first constructing a new collection object and then recursively populating it with copies of the child objects found in the original. In case of deep copy, a copy of object is copied in other object.

```python
from copy import copy , deepcopy
  
# initializing list 1 
l1 = [1, 2,'hello','rankush']

# using copy for shallow copy  
l2 = copy(l1) 
  
# using deepcopy for deepcopy  
l3 = deepcopy(l1)
```

```python
# using copy
l2[3] = 3
l1[1] = 'one'
print(l3, l2, l1)
```

```python
# using deep copy
l3[2] = 'ram'
l1[1] = 1
print(l3, l2, l1)
```
