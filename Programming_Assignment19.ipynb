```python
def double_char(str):
  return ''.join([i*2 for i in str])

double_char('String')
```

```python
double_char('Hello World!')
```

```python
def reverse(b):
  if type(b) == bool:
    return not b
  else:
    return 'boolean expected'

reverse(True)
```

```python
reverse(False)
```

```python
reverse(0)
```

```python
reverse(None)
```

```python
def num_layers(n):
  res = 0.5
  for _ in range(n):
      res *= 2
  return res / 1000 # for meters

num_layers(1)
```

```python
num_layers(4)
```

```python
num_layers(21)
```

```python
def index_of_caps(str):
  index = []
  for i in range(len(str)):
    if str[i].isupper():
      index.append(i)
  return index

index_of_caps('eDaBiT')
```

```python
index_of_caps('sUn')
```

```python
index_of_caps('eQuINoX')
```

```python
def find_even_nums(n):
  return [i for i in range(1,n+1) if i%2==0]

find_even_nums(8)
```

```python
find_even_nums(4)
```

```python
find_even_nums(2)
```
