```python
import math
def Question1(args,C = 50, H=30):
  for i in args:
    print(int(math.sqrt((2*i*C)/H)), sep = ',', end=' ')

Question1(list(map(int,input().split(','))))
```

```python
def Question2(X,Y):
  _list = [[0 for col in range(Y)] for row in range(X)]

  for row in range(X):
      for col in range(Y):
          _list[row][col]= row*col
  
  return _list
```

```python
Question2(3,5)
```

```python
def Question4(lst):
  result = []
  lst.sort()
  for i in lst:
    if i not in result:
      result.append(i)
  return ' '.join(result)

Question4(input().split())
```

```python
def Question5(sent):
  letters = 0
  digits = 0
  for i in sent:
    if ord(i) in range(65,91) or ord(i) in range(97,123):
      letters += 1
    elif ord(i) in range(48,58):
      digits += 1
    else:
      pass
  return f'LETTERS {letters} \nDIGITS {digits}'
```

```python
print(Question5(input()))
```

```python
def Valid(pass_):
  special  = '$#@'
  if len(pass_) >= 6 and len(pass_) <= 12 and any(ch.isdigit() for ch in pass_) and any(ch.isupper() for ch in pass_) and any(ch.islower() for ch in pass_) and any(ch in special for ch in pass_) :
    return True
  else:
    return False
def Question6(lst):
  for i in lst:
    if Valid(i):
      print(i, end=',')

```

```python
Question6(input().split(','))
```
