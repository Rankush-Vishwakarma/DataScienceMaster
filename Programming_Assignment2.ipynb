```python
# Q.1
def KM_to_Miles(km):
  # 1 km = 0.621371 miles
  return km * 0.621371

KM_to_Miles(9)
```

```python
# Q.2
def Cel_to_Faren(cel):
  # 1 celcius = 33.8 fahrenheit
  return cel * 33.8

Cel_to_Faren(40)
```

```python
# Q.4 Solve the quadratic equation 
import cmath

def SolveQuad(a,b,c):
  # calculate the discriminant
  d = (b**2) - (4*a*c)

  first = (-b-cmath.sqrt(d))/(2*a)
  second = (-b+cmath.sqrt(d))/(2*a)

  print(f'{a}x^2 {b}x {c} = {first} and {second}')
```

```python
SolveQuad(1,2,3)
```

```python
def swap(a,b):
  a = a+b
  b = a-b
  a = a-b
  return a,b

swap(5,4)
```
