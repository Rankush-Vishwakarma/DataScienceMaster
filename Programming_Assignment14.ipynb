```python
class Question1:
  def __init__(self,n):
    self.n = n
  def gen(self):
    for i in range(self.n):
      if i % 7 == 0:
        yield i

ques = Question1(30)
gen = ques.gen()
next(gen)
```

```python
def returnFreq(str):
  list_ = str.split()
  dict_ = {}
  for i in list_:
    if i in dict_:
      dict_[i] += 1
    else:
      dict_[i] = 1
  return dict_ 

returnFreq(input())
```

```python
def MakeSentence(sub, verb, obj):
  sentenceList = [s+' '+v+' '+o for s in sub for v in verb for o in obj]
  for i in sentenceList:
    print(i)

MakeSentence(['I',"You"],["Play","Love"],['Hockey','Football'])
```

```python
import gzip
def Compress(s):
  s = bytes(s, 'utf-8')
  return gzip.compress(s)

def Decompress(s):
  return gzip.decompress(s).decode('utf-8')
```

```python
s = Compress('Hello world!helloworld!hello world!hello world!')
```

```python
print(s)
```

```python
Decompress(s)
```

```python
def binarySearch(lst, value):
  low = 0
  high = len(lst) - 1
  mid = 0

  while low <= high:

      mid = (high + low) // 2

      # If x is greater, ignore left half
      if lst[mid] < value:
          low = mid + 1

      # If x is smaller, ignore right half
      elif lst[mid] > value:
          high = mid - 1

      # means x is present at mid
      else:
          return mid

  # If we reach here, then the element was not present
  return -1
```

```python
binarySearch([1,2,3,4,5,6,7,8,9], value = 6 )
```
