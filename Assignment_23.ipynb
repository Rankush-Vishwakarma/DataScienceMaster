```python
def func(a,b,c=5):
  print(a,b,c)

func(1,c=3,b=2)
```

This style of passing parameter is known as keyword arguments, in this case we don't have to ch care about the order the parameter to be passed. we can replace the position of the parameter with their respective keywords. like here c is place at last but we can position it to the middle and update the respective arguments.

image.png

```python
def func(a, *pargs):
  print(a, pargs)

func(1,2,3)
```

this style of passing parameter is known as arbitary keywords, this can be very helpful in case if we don't know how many parameter we need to pass, in this case we have pass one must keyword and rest we don't know how many we can have hence we are useing *pargs , *pargs can have any number of values, of any data types

```python
def func(a,**kargs):
  print(a,kargs)

func(a=1, c=3, b=2)
```

If we want to pass a keyworded variable length of arguments to a function, we use **kwargs.
