```python
for i in range(10):
  print(f'hii-{i}', end=' ')
print('\n')
for i in range(0,10):
  print(f'hii-{i}', end=' ')
print('\n')
for i in range(0,10,1):
  print(f'hii-{i}', end=' ')    
```

We can clearly see that the output of all the three loops are same, basically range is an inbuild function in which by default the starting index is always set to 0 if we are not providing it and the increametation step by value is always 1.
> range(start, stop, step), by default start = 0 and step =1.
