```python
# Q. 1

def fib(num):
  if num <= 1:
    return num
  else:
    return(fib(num-1) + fib(num-2))


for i in range(10):
  print(fib(i),end=' ')
```

```python
#  Q . 2

def fact(n):
  if n <= 1:
    return 1
  else:
    return n * fact(n-1)

fact(5)
```
