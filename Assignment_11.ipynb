In python the five levels of logging is defined in order of increasing the severity of the code. these are:


1.   DEBUG
2.   INFO
3.   WARNING
4.   ERROR
5.   CRITICAL

Print statement work for real time debugging but when it comes to long term running software it become very difficult for any tester to read all the print statement line by line instead he/she can directly refer the logger file .a logger is ideal for any time you need to debug the script after it has failed rather than while it is failing.

Step Over: 	If the current line contains a function call, Step Over runs the code and then suspends execution at the first line of code after the called function returns.
Step Out:	Step Out continues running code and suspends execution when the current function returns. The debugger skips through the current function.
Step in : An action to take in the debugger. If the line does not contain a function it behaves the same as “step over” but if it does the debugger will enter the called function and continue line-by-line debugging there.
