```python
def AddMatrix(m1, m2):
  # addition or subtraction can only happen in matrics if the shape are equal
  if len(m1) == len(m2) and len(m1[0]) == len(m2[0]):
    for i in range(len(m1)):
      for j in range(len(m1[0])):
        m1[i][j] += m2[i][j]
    return m1
  else:
    return 'False, Inconsistant shape'
  

AddMatrix([[1,2],[3,4]],[[4,5],[6,7]])
```

```python
AddMatrix([[1,2],[3,4]],[[4,5,6],[6,7]])
```

```python
def MulMatrics(m1, m2):
  return [[sum(a * b for a, b in zip(A_row, B_col))
                        for B_col in zip(*m2)]
                                for A_row in m1]

MulMatrics([[1,2],[3,4]],[[4,5],[6,7]])
```

```python
def Transpose(m):
  return [[m[j][i] for j in range(len(m))] for i in range(len(m[0]))]

Transpose([[1,2],[3,4]])
```

```python
def SortAlphabetic(arr):
  # using insertion sort
  for i in range(2, len(arr)):
        key = ord(arr[i])
        # Moving elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i-1
        while j >= 0 and key < ord(arr[j]) :
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = chr(key)
  return arr
```

```python
SortAlphabetic(['B','D','A','E','C'])
```

```python
# word sorting
def sortit(sentence):
  word =  [word.lower() for word in sentence.split()]
  word.sort()
  return word

print(sortit('Hello sir , I need job  please help me'))
```

```python
# remove punctuation
import string
punc = string.punctuation

def removePuncuations(string):
  new = []
  for i in string:
    if i not in punc:
      new.append(i)
  return ''.join(new)


removePuncuations('hello!everyone@i-need$a#job&please.help.Me')
```
