```python
def filter_list(lst):
  return [i for i in lst if type(i) == int]
```

```python
filter_list([1,2,'a','b'])
```

```python
filter_list([1,'a','b',0,15])
```

```python
def fact(n):
  if n < 1:
    return 1
  else:
    return n * fact(n-1)

fact(5)
```

```python
def move_to_end(lst, value):
  for i in range(len(lst)):
    if lst[i] == value:
      lst.append(lst.pop(lst.index(lst[i])))
  return lst

move_to_end([1,3,2,4,4,1],1)
```

```python
move_to_end(['a','a','a','b'],'a')
```
