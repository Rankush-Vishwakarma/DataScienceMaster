````python
# -*- coding: utf-8 -*-
"""
Created on Sat Feb 12 20:56:04 2022

@author: rvishwakar23
"""

import os
import shutil

def fileCopier(filetype,src, destination):
    print(f"Please wait i'm loading the {filetype} ")
    fileList = []
    # getting the default base directory path
    dir_path = os.path.expanduser('~')
    for root, _, files in os.walk(dir_path):
        if len( root.split('\\'))>3 and root.split('\\')[3] in src:
            for file in files:
                #print(file)
                if file.endswith(f'.{filetype}'):
                    print(file)
                    filePath = root+'\\'+str(file)
                    fileList.append(filePath)
                else:
                    pass
    try:
        for f in fileList:
            shutil.copy(f,dir_path+'\\'+ destination)
        print(f'file copied from {src} to {destination} successfully.')
    except Exception as e:
        print(f'Not able to copy the file {str(e)}')
````
