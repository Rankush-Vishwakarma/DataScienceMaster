```python
# Q. 1
import numpy as np
def sum(l):
  array = np.array(l)
  return np.sum(array)

sum([1,2,3,4,5,6,7,8])
```

```python
# Q. 2
def max(l):
  array = np.array(l)
  return np.max(array)

max([1,2,3,2,10,23,344])
```

```python
def rotate(l):
  newarray = []
  for i in range(1,len(l)+1):
    newarray.append(l[-i])
  return newarray

rotate([1,2,3,4,5,6,7,8,9])
```

```python
def solve4(arr):
  length  = len(arr)
  mid = length//2
  first = arr[:mid]
  second = arr[mid+1:]
  return  second + first


print(solve4([1,2,3,4,5,6]))
```

```python
def checkMonotonic(arr):
  return (all(arr[i] <= arr[i + 1] for i in range(len(arr) - 1)) or
            all(arr[i] >= arr[i + 1] for i in range(len(arr) - 1)))
  
checkMonotonic([1,3,0,4])
```

```python
checkMonotonic([1,3,4,5])
```
