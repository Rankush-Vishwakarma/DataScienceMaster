InputInt(): - used to take integer value inout this will also tae some additional parameter like minimum value ('min'), maximum value (max) , greater than value (greaterThan) and less than value (lessThan) and return the particular integer number.


inputFloat(): -used to Accpet the floating point numerical value and the same as in inputInt() it also take 4 additional parameters like minimum value ('min'), maximum value (max) , greater than value (greaterThan) and less than value (lessThan) and return the particular float number

```python
import pyinputplus as pyip
# in this example i have taken the minimum value as 1 and the value less than 10 
intInput = pyip.inputInt(prompt='Enter the number', min=1,lessThan=10) 
print(intInput)
```

```python
floatInput = pyip.inputFloat(prompt='Enter the number', min=1,lessThan=10)
print(floatInput)
```

AllowRegexes parameter allows to set a list of regex string eith any pyinputplus method. A sequence of regex str that will explicitly pass validation.

blocklistRegexes parameter allows A sequence of regex str or (regex_str, error_msg_str) tuples that, if matched, will explicitly fail validation.

```python
phone = pyip.inputNum(prompt='What is your phone number?',  allowRegexes=[r'\d{3}-\d{3}-\d{4}', r'None'])
print('Your phone is - ', phone)
```

```python
phone = pyip.inputNum(prompt='What is your phone number?',  blockRegexes=[r'\d{3}-\d{3}-\d{4}', r'None'])
print('Your phone is - ', phone)
```
