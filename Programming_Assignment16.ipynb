```python
def stutter(str):
  if len(str) < 2:
    return 'not having enough length of string to st... st... stutter'
  else:
    return f"{(str[:2]+'... ')*2 + str+'?'}"

stutter('incredible')
```

```python
stutter('enthusiastic')
```

```python
stutter('outstanding')
```

```python
def area_of_hexagon(n):
  return round((3 * (3**0.5) * n**2) / 2,1)

area_of_hexagon(1)
```

```python
area_of_hexagon(2)
```

```python
area_of_hexagon(3)
```
