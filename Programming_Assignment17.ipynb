```python
def evenly_divisible(a,b,c):
  return sum([i for i in range(a,b+1) if i%c==0 ])

evenly_divisible(1,10,20)
```

```python
evenly_divisible(1,10,2)
```

```python
evenly_divisible(1,10,3)
```

```python
def replace_vowels(str, ch):
  byte = list(str)
  for i in range(len(byte)):
    if byte[i] in 'aeiou':
      byte[i] = ch
  return ''.join(byte)

replace_vowels('the aardvark','#')
```

```python
replace_vowels('minnie mouse','?')
```

```python
replace_vowels('shakespeare','*')
```

```python
def fact(n):
  if n <= 1:
    return 1
  else:
    return n * fact(n-1)
```

```python
fact(5)
```

```python
fact(3)
```

```python
fact(1)
```

```python
fact(0)
```

```python
def hamming_distance(s1, s2):
  count = 0
  if len(s1) == len(s2):
    for i in range(len(s1)):
      if s1[i] != s2[i]:
        count+=1
  if len(s1) > len(s2):
    for i in range(len(s2)):
      if s2[i] != s1[i]:
        count+=1
    count = count + (len(s1)-len(s2))
  if len(s1) < len(s2):
    for i in range(len(s1)):
      if s1[i] != s2[i]:
        count+=1
    count = count + (len(s2)-len(s1))
      


  return count
```

```python
hamming_distance('strong','strung')
```

```python
hamming_distance('abcde','bcdef')
```

```python
hamming_distance('abcde','abcde')
```
