```python
def FindGreater(lst, k):
  for i in lst:
    if len(i) >= k:
      print(i, end = ' ')
    else:
      pass

FindGreater(['I','am','rankush','vishwakaram','I','am','python programmer','please','help','me','getting','job'], 3)
```

```python
def Removeithcharacter(str, k):
   before = str[:k]
   after = str[k+1:]
   return before + after

Removeithcharacter('python programmer',6)
```

```python
def SplitJoin(str):
  split = str.split()
  return '_'.join(split)

SplitJoin('Write a Python program to split and join a string')
```

```python
# we can do the same with the help of set set has no duplicate elements
def UncommonWords(s1,s2):
  # uncommon from first to second string
  for i in range(len(s1)):
    if s1[i] not in s2:
      print(s1[i], end=' ')
  # uncommon from second to first string
  for i in range(len(s2)):
    if s2[i] not in s1:
      print(s2[i], end=' ')

UncommonWords('rankush','vishwakarma')
```

```python
def FindDuplicate(str):
  set_ = set()
  for i in range(len(str)):
    if str[i] in str[i+1:]:
      set_.add(str[i])
  return set_

FindDuplicate('vishwakarmakumar')
```

```python
def CheckSpecialChar(str):
  spcialChar = '!@#%^&*'
  for k in str:
    if k in spcialChar:
      return True
      break
  return False

CheckSpecialChar('rankush.vishwakarama@gmail.com')
```
