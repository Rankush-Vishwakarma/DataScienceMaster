```python
# Q.1

def fact(n):
  if n <= 1:
    return 1
  else:
    return n * fact(n-1)

fact(5)
```

```python
# Q. 2
def multiplication(num):
  for i in range(1, 11):
    print(f'{num}\t*\t{i}\t=\t{num * i}')

multiplication(4)
```

```python
# Q.3
def fib(n):
  if n < 0:
        print("Incorrect input")
 
  # Check if n is 0
  # then it will return 0
  elif n == 0:
      return 0
      
  elif n == 1 or n == 2:
      return 1

  else:
      return fib(n-1) + fib(n-2)


print(fib(10))
```

```python
# Q. 4
def armstrongCheck(num):
  sum = 0
  org_num = num
  while org_num > 0:
   digit = org_num % 10
   sum += digit ** 3
   org_num //= 10
  return num == sum

armstrongCheck(153)
```

```python
# Q. 5

def sum(num):
  sum = 0
  for i in range(num):
    sum += i
  return sum

sum(30)
```
