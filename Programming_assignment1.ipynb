```python
#  Q.2
def AddDiv(x,y):
  opr = int(input('1: add \n2: divide \n'))
  if opr == 1:
    return f'addition of {x} + {y} = {x + y}'
  elif opr == 2:
    return f'division of {x} / {y} = {x/2}'
  else:
    pass

AddDiv(4,5)
```

```python
# Q. 3
import math
def AreaOfTriangle(x,y,z):
  p = (x+y+z)//2
  return f'Area of triangle with {x},{y} and {z} sides = {math.sqrt(p*(p-x)*(p-y)*(p-z))}'
```

```python
AreaOfTriangle(24,30,18)
```

```python
# Q. 5
from time import time
def Random(min, max): 
  t_ = time()
  avg = (min+max)//2
  num1 = t_%avg-min
  num2 = t_%avg+min
  if int(num1) in (list(range(min,max))):
    return int(num1)
  elif int(num2) in (list(range(min,max))):
    return int(num2)
  elif num1 > min :
    return int(num1)
  elif num2 < max :
    return int(num2)
  else:
    return avg

  
```

```python
for i in range(10):
  print(Random(i,7))
```
