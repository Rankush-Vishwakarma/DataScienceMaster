The simplest way to save a programme from crashing it to use Try and except cluase block , this will try to catch the error and further we can use pass keyword in order to simply pass the function without having any difficulty

```python
def divide_by_zero(num):
  try:
    return num//0
  except Exception as e:
    print(e)
    pass
```

```python
divide_by_zero(3)
```
