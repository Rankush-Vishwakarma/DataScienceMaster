# using def

````python
# lets understand with some example

'''
The printSquare function will take one integer type of value and will return the
square value of it. Here we have to use namespace 'printSquare' in order to 
define the function and we are trying to return some integer type of value from it.
'''
def printSquare(x):
  return x**2

````

````python
'''
In order to get the square root of a value we must have to using the defined 
namespace of the function and pass the desired argument value into it, and thus 
we will get the square value of it.
'''
print(printSquare(4))
````

# using lembda function

````python
"""
Here we are doing the same thing but 'printSquare' is a variable that hold an 
expression and the expression is basically calculating the value. in lamda by 
defult we need to return some value , but in def if we don't want to return 
we can simple print or pass but in lembda function we cannot do this.
lambda function doesn't include any return statement.
"""
printSquare = lambda x: x**2
````

````python
'''
Calling lambda function and def keyword is the same way, we need to use the expression
Variable to use the value.
'''
printSquare(4)
````

## filter()
-- filter() forms a new list that contains only elements that satisfy a certain condition, i.e. the function we passed returns True or False depanding upon the condition defined to be satisfy.


```
# filter(function, iterable(s))
```

### lets take an string type of number as input and map it to the integer value.

````python
string_int_input = input().split() # this has list of string type
````

````python
print(string_int_input)
````

````python
# lets map these numbers to intger type . so we will use int function to convert it
# since map function will return the map object we are converting it to list.
new_input_list = list(map(int, string_int_input))
````

````python
# lets defnine a list with multiple data type value
list_ = [1,'A',2,'R','B',3,4,5,'K']
````

>>> using traditional approch

````python
# traditional approch using function
def add(l):
  sum = 0
  for i in range(len(l)):
    sum+=l[i]
  return sum

````

````python
add([1,2,3,4,5,6,7,8])
````

````python
# using reduce
from functools import reduce 
def add(x,y):
  return x+y
````

````python
reduce(add, [1,2,3,4,5,6,7,8])
````

## working 
### reduce take function as first argument, an iterable object i.e., sequence  and the optional intial position. in tradition approch we are doing the same thing basically we are adding first value with the 0 and the result of this will be added with another value of the sequence and the loop keeps going until the end of the sequence.

### in reduce function we are doing the same thing we are adding first two values add(1,2) - 3 and add(3,4)-7 and process repeats until we've gone through all the elements in the sequence.

Function annotation is a PEP(python enhancement Proposal) standard for arbitary metadata annotation to python writing style. These are the expression that used for various part of functions. it is basically evaluated at the compilation time and having no life in python runtime environment.

Syntax:
```
#  identifier [: expression] [= expression].
for example:
def fib(n:'int', output:'list'=[])-> 'list' ...
```
the annotation cannot be used with lambda. it doesn't support the annotation.
The annotation can be available via function.__annotions__ this will returns the key value pair of annotation that has been used .

Recursive functions are those function which call itself until it reach to some condition and calculate the desired result.

````python
# lets find the sum of the nummbers using the recursion

def findSumUsingRecursion(number:'int') -> int:
  if number <= 1:
    return number 
  else:
    return number + findSumUsingRecursion(number-1)
````

````python
findSumUsingRecursion(5)
````

1. Naming Conventions - Use a lowercase word or words. Separate words by underscores to improve readability.
2. Code Layout - Surround top-level functions and classes with two blank lines Surround method definitions inside classes with a single blank line, Use blank lines sparingly inside functions to show clear steps.
3. Maximum Line Length and Line Breaking - PEP 8 suggests lines should be limited to 79 characters. in function there should not be more than 79 character in the paramter of the function or inside the function.
4. Indentation - Use 4 consecutive spaces to indicate indentation.
5. Line up the closing brace with the first non-whitespace character of the previous line.
6. Limit the line length of comments and docstrings to 72 characters, Use complete sentences, starting with a capital letter, Make sure to update comments if you change your code.
7. Surround docstrings with three double quotes on either side, as in """This is a docstring""".
