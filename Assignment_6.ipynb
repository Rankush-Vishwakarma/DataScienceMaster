An escape character is a backslash "\ "followed by the character that need to insert. Escape charcter is used in the place where it is difficult for inserting the invalid character or string in an string.

```python
txt = 'hello i am 'rankush' and i am data scientist'
```

The above text is invalid because '' 

```python
txt = "hello i am \"rankush\" and i am data scientist"
```
