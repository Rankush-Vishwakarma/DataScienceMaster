````python
import re
regex = re.compile(r'(\d\d\d)-(\d\d\d-\d\d\d\d)')
mo = regex.search('My number is 91-825-301-5353.')
print('group zero- ',mo.group(0))
print('group one- ',mo.group(1))
print('group two- ',mo.group(2))
````

'\w'	Returns a match where the string contains any word characters (characters from a to Z, digits from 0-9, and the underscore _ character).
'\s'	Returns a match where the string contains a white space character

'\W'	Returns a match where the string DOES NOT contain any word characters.

re.VERBOSE allows the user to write the regular expression that look nicer and more readable by allowing you to visually separate logical section of the pattern and add comments.

````python
# when we used to write without using verbose
regex_email = re.compile(r'^([a-z0-9_\.-]+)@([0-9a-z\.-]+)\.([a-z\.]{2, 6})$',
              re.IGNORECASE)

# with verbose we can write the same pattern in more readable and beaultiful ways.
regex_email = re.compile(r"""
            ^([a-z0-9_\.-]+)              # local Part
            @                             # single @ sign
            ([0-9a-z\.-]+)                # Domain name
            \.                            # single Dot .
            ([a-z]{2,6})$                 # Top level Domain  
             """,re.VERBOSE | re.IGNORECASE)  
````

````python
name_regex = re.compile(r'[A-Z](?:\w)+\sWatanabe')
list_ = ['Haruto Watanabe','Alice Watanabe','RoboCop Watanabe', 'Watanabe','Mr. Watanabe','Haruto watanabe']
for i in list_:
  print(name_regex.search(i))
````

Lets break `'[A-Z](?:\w)+\sWatanabe'` the first part `[A-Z](?:\w)` will check for first name and `[A-Z]` would try to validate only for first name having first letter as capital letter and then followed by some other character and after that a space and `\sWatanabe` check for that last name is exactly math the particular string.
