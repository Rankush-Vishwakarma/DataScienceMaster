```python
def is_symmetrical(n):
  return str(n)==str(n)[::-1]
```

```python
is_symmetrical(7227)
```

```python
is_symmetrical(12567)
```

```python
is_symmetrical(444444444)
```

```python

def multiply_nums(nums):
  prod = 1
  nums = list(map(int, nums.split(', ')))
  for  i in nums:
    prod *= i
  return prod
multiply_nums("2, 3")
```

```python
multiply_nums("1, 2, 3, 4")
```

```python
def square_digits(nums):
  prod = []
  for digits in str(nums):
    prod.append(str(int(digits)**2))
  return int(''.join(prod))
```

```python
square_digits(9119)
```

```python
square_digits(2483)
```

```python
square_digits(3212)
```

```python
def mean(num):
  sum = 0
  for i in str(num):
    sum += int(i)
  return sum // len(str(num))
```

```python
mean(42)
```

```python
mean(12345)
```

```python
mean(666)
```
