```python
X = 'iNeuron'
def func():
  X  = 'Nl'
  def nested():
    print(X)
  nested()
```

```python
func()
X
```

In above code nested method concept are getting use which means, there is a main function or method and inside that there is another function , in these scenario the value and function has local scope hence calling func() function would defind another local function nested and print the value of X ** key thing is the print(X) state take X value as local which is defined inside the func() function hence for nested function it is treated as global variable but for func() it is treated as local and function nested is also treated as local. but irrespective of local X The Globla variable X value remain safe as there is no change is going to happen globally.
