```python
# Q. 1
def check(num):
  if num > 0 :
    return 'Positive'
  elif num < 0:
    return 'Negative'
  elif num == 0:
    return 'Zero'
  else:
    pass

print(check(3))
print(check(0))
print(check(-4))
```

```python
# Q. 2
def EveOdd(num):
  if num%2 == 0:
    return 'Even'
  else:
    return 'Odd'

EveOdd(3)
```

```python
# Q.3
def leapYear(year):
  return (year % 400 == 0) and (year % 100 == 0)

leapYear(2000)
                               
```

```python
# Q . 4
from math import sqrt
def primeCheck(num):
    if (num <= 1): 
      return False
    for i in range(2, int(sqrt(num))+1):
      if (num % i == 0):
          return False
    return True

primeCheck(3)
```

```python
def prime(min, max):
  prime = []
  for number in range (min, max + 1):  
    if number > 1:  
        for i in range (2, int(sqrt(number))+1):  
            if (number % i) == 0:  
                break  
        else:  
            prime.append(number)
  return prime
```

```python
prime(1,1000)
```
