```python
# Q. 1 - Program to find the LCM of two numbers

def LCM(x, y):
   # choose the greater number
   if x > y:
       greater = x
   else:
       greater = y

   while(True):
       if((greater % x == 0) and (greater % y == 0)):
           lcm = greater
           break
       greater += 1
   return lcm

LCM(40,30)
```

```python
# Q. 2 program to find HCF of two numbers

def HCF(x, y):
    # choose the smaller number
    if x > y:
        smaller = y
    else:
        smaller = x

    for i in range(1, smaller+1):
        if((x % i == 0) and (y % i == 0)):
            hcf = i 
    return hcf

HCF(40,30)
```

```python
# Q. 3 - using predefined function
def BinOctHex(num):
  return f'binary - {bin(num)}, octal - {oct(num)}, hexadecimal - {hex(num)}'
BinOctHex(144)
```

```python
# user defined 
bin  = []
oct = []
hex = []

def binary(num): 
  if num >= 1:
    binary(num // 2)
  bin.append(str(num % 2))

def octal(num): 
  if num >= 1:
    octal(num // 8)
  oct.append(str(num % 8))

def hexa(num):
  if num >= 1:
    hexa(num // 16)
  hex.append(str(num % 16))

def BinOctHex(num):
  binary(num)
  octal(num)
  hexa(num)
  bin_ = ''.join(bin)
  oct_ = ''.join(oct)
  hex_ = ''.join(hex)

  return f'binary - {bin_}, octal - {oct_}, hexadecimal - {hex_}'


BinOctHex(144)
```

```python
from functools import reduce
def calculator(*args):
  opr = int(input('1. Addition\n2. Subtraction\n3. Multiplication\n4.Division\n\n'))
  if opr == 1:
    print(sum(args))
  elif opr == 2:
    print(args[0] - args[1])
  elif opr == 3:
    print(reduce((lambda x, y: x * y), args))
  elif opr == 4:
    print(args[0] / args[1])
  else:
    return 'wrong option'
```

```python
calculator(1,2)
```
